```python
import pandas as pd
df = pd.read_csv("Hindi_clean.csv", encoding="utf-8")
```

```python
pd.set_option('display.max_colwidth', None)   # show full text
pd.set_option('display.max_rows', 100) 
df.head(30)
```

```python
def clean1(text):
    import re, html
    
    text = html.unescape(text)
    
    # remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # remove mentions
    text = re.sub(r'@\w+', '', text)
    
    # remove RT
    text = re.sub(r'\bRT\b', '', text)
    
    # remove leading junk
    text = re.sub(r'^[:\-\s]+', '', text)
    
    # remove symbols like | ()
    text = re.sub(r'[|()]+', '', text)
    
    # remove ellipsis (...)
    text = re.sub(r'\.{2,}', ' ', text)
    
    # normalize spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
df['text']= df['text'].apply(clean1)
```

```python
import re

def final_test_prep(text):
    if not isinstance(text, str): return ""
    
    # 1. Strip leading/trailing quotes and colons from CSV export
    text = text.strip('": ')
    
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    
    # 3. Strip '#' but keep the word
    text = re.sub(r'#(\w+)', r'\1', text)
    
    # 4. Remove extra whitespace and newlines
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

df['clean_text'] = df['text'].apply(final_test_prep)
```
