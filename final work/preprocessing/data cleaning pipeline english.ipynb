```python
import re
import html

def clean_english_final(text):
    text = html.unescape(text)
    
    # remove RT + mentions
    text = re.sub(r'\bRT\b', '', text)
    text = re.sub(r'@\w+', '', text)
    
    # remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # reduce repeated punctuation
    text = re.sub(r'!{2,}', '!', text)
    text = re.sub(r'\?{2,}', '?', text)
    
    # remove leading junk (!!!, :, spaces)
    text = re.sub(r'^[:\-\s!]+', '', text)
    
    # remove ellipsis
    text = re.sub(r'\.{2,}', ' ', text)
    
    # remove quotes (normal + fancy)
    text = re.sub(r'["“”‘’]', '', text)
    
    # normalize spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
df['clean_text'] = df['clean_text'].apply(clean_english_final)
```

```python
import pandas as pd
import re

def preprocess_text(text):
    if not isinstance(text, str):
        return ""

    # 1. Remove leading/trailing special characters (!!!!, quotes, colons, spaces)
    # This regex looks at the start (^) and end ($) of the string
    text = re.sub(r'^[!:\s"\'“”]+', '', text) 
    text = re.sub(r'[!:\s"\'“”]+$', '', text)

    # 2. Reduce multiple exclamation/question marks to a single one
    # This keeps the 'intent' but removes the noise
    text = re.sub(r'!+', '!', text)
    text = re.sub(r'\?+', '?', text)

    # 3. Remove extra whitespace
    text = " ".join(text.split())

    return text

# Assuming your dataframe is named 'df'
df['clean_text'] = df['clean_text'].apply(preprocess_text)

# Preview results
print(df['clean_text'].head(10))
```

```python
import pandas as pd
import re

def advanced_clean(text):
    if not isinstance(text, str):
        return ""

    # 1. Remove the specific leading/trailing junk from your CSV export
    # This handles: "Initial quotes, !!!!, and leading spaces
    text = re.sub(r'^[!:\s"\'“]+', '', text)
    text = re.sub(r'[!:\s"\'”]+$', '', text)

    # 2. Fix the "double quote" artifacts inside the text
    # Often happens when CSVs export text containing quotes
    text = text.replace('""', '"')

    # 3. Strip the '#' from hashtags but keep the word
    text = re.sub(r'#(\w+)', r'\1', text)

    # 4. Remove URLs (Hate speech models don't need links)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # 5. Handle repetition (Optional but recommended)
    # Reduces "LMAOOOOO" to "LMAOO" so the tokenizer doesn't get confused
    text = re.sub(r'(.)\1{3,}', r'\1\1', text)

    # 6. Final whitespace cleanup
    text = " ".join(text.split())
    
    return text


df['clean_text'] = df['clean_text'].apply(advanced_clean)
```
